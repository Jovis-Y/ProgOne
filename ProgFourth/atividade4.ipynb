```python
def menu():
  print("=" * 40)
  print("LISTA DE CONTATOS".center(40, " "))
  print("=" * 40)
  print("[0]: encerrar")
  print("[1]: incluir novo nome")
  print("[2]: incluir telefone")
  print("[3]: excluir telefone")
  print("[4]: excluir nome")
  print("[5]: consultar telefone")

def incluirNovoNome(nome, numeros):
  global listaNumeros
  listaNumeros[nome] = numeros

def incluirTelefone(nome, numero):
  global listaNumeros
  
  if nome in listaNumeros.keys():
    listaNumeros[nome].append(numero)
  elif input("Nome não encontrado, deseja adicionar o nome e número? [1 para confirmar]: ") == "1":
    incluirNovoNome(nome, numero)

def excluirTelefone(nome, numero):
  global listaNumeros
  
  if len(listaNumeros[nome]) == 1:
    del listaNumeros[nome][0]
  elif len(listaNumeros[nome]) > 1:
    listaNumeros[nome].remove(numero)

def excluirNome(nome):
  global listaNumeros

  listaNumeros.pop(nome, None)

def consultarTelefone(nome):
  global listaNumeros

  print(" | ".join(listaNumeros.get(nome, " ")))

while True:
  menu()
  opcao = int(input("Escolha uma opção: "))
  if opcao == 0:
    break
  nome = input("Insira o nome: ")

  if opcao == 1:
    quantidade = int(input("Quantos números deseja adicionar: "))
    numeros = [input("Insira o número: ") for n in range(quantidade)]
    incluirNovoNome(nome, numeros)

  elif opcao == 2:
    numero = input("Insira o número: ")
    incluirTelefone(nome, numero)
  
  elif opcao == 3:
    numero = input("Insira o número: ")
    excluirTelefone(nome, numero)
  
  elif opcao == 4:
    excluirNome(nome)
  
  elif opcao == 5:
    consultarTelefone(nome)
```

```python
corredores = dict()

for n in range(6):
  nome = input("Insira o nome: ")
  tempo = [int(input(f"Insira o tempo da {i}° volta: ")) for i in range(1, 11)]
  corredores[nome] = tempo

minimo = min([sum(time)/10 for time in corredores.values()])
print("=" * 40)
print("Corredores: média")
for nome, tempo in corredores.items():
  msg = f"{nome}: {sum(tempo)/10:.2f}"
  print(msg)
  if sum(tempo)/10 == minimo:
    vencedor = msg
print("=" * 40)

print(f"Vencedor = {vencedor}")
```

```python
tabelaConsumoEner = {
  'ar-condicionado':1600,
  'computador':     350,
  'chuveiro':       5000,
  'ferro':          1000,
  'lampada':        32,
  'lavadora-roupas':600,
  'refrigerador':   350,
  'tv':             200
}

tabelaBandeira = {
    'verde':   0.50,
    'amarela': 0.53,
    'vermelha':0.56
}

horasEquipamento = dict()
bandeira = input("Insira a bandeira[verde, amarela, vermelha]: ")
ICMS = 1 + float(input("Insira a taxa do ICMS: "))
quantidade = int(input("Insira a quantidade de equipamentos: "))

for n in range(quantidade):
  equipamento, consumo = input(f"Insira o nome do equipamento e as horas de uso[separados por um espaço]: ").split()
  horasEquipamento[equipamento] = float(consumo.replace(",", "."))

conta = sum([tabelaConsumoEner[equip] * horasEquipamento[equip] / 1000 
             for equip in horasEquipamento.keys()]) * ICMS * tabelaBandeira[bandeira]
print(f"valor da conta:", conta)


```

```python
dadosLutadores = dict()
lutadoresAprovados = dict()
pesoInferior, pesoSuperior = input("Insira o peso limite inferior e superior: ").split()
quantidade = int(input("Insira a quantidade de lutadores: "))

for n in range(quantidade):
  lutador, peso = input("Insira o nome e peso do lutador: ").split()
  dadosLutadores[lutador] = int(peso)

for nome, peso in dadosLutadores.items():
  if int(pesoInferior) <= peso <= int(pesoSuperior) * 1.08:
    lutadoresAprovados[nome] = peso

print("\nquantidade de lutadores aprovados:", len(lutadoresAprovados))
print(f"porcentagem de reprovação: {100 - len(lutadoresAprovados) / len(dadosLutadores) * 100:.2f}%")
if len(lutadoresAprovados) == 0:
  print('vazio')
else:
  for nome, peso in lutadoresAprovados.items():
    if peso == min(lutadoresAprovados.values()):
      print("lutador aprovado com menor peso:",nome)
```
